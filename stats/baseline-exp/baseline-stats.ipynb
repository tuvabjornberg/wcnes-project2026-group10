```python
import matplotlib.pyplot as plt

distance = [50, 100, 150, 200, 250, 275]
ber = [19.31, 10.01, 20.21, 16.07, 28.04, 30.80]
pdr = [100, 100, 100, 100, 2.5, 2.0]
rssi = [-99.53, -98.42, -100.67, -100.11, -100.40, -102.75]
rate = [6.62, 26.53, 4.43, 17.15, 2.66, 0.19]

distance, ber = zip(*sorted(zip(distance, ber)))
```

### BER-Distance

```python
plt.figure()
plt.bar(distance, ber, width=20)
plt.xlabel("Distance [cm]")
plt.ylabel("Bit Error Rate [%]")
plt.title("Baseline BER vs Distance")
plt.grid(axis='y')
plt.show()
```

### PDR-Distance

```python
plt.figure()
plt.plot(distance, pdr, marker='o')
plt.xlabel("Distance [cm]")
plt.ylabel("PDR [%]")
plt.title("Packet Delivery Ratio vs Distance")
plt.grid()
plt.xticks(distance)
plt.show()
```

### PER-Distance??

```python
per = [100 - x for x in pdr]

plt.figure()
plt.plot(distance, per, marker='o')
plt.xlabel("Distance [cm]")
plt.ylabel("PER [%]")
plt.title("Packet Error Rate vs Distance")
plt.grid()
plt.xticks(distance)
plt.show()
```

### RSSI-Distance

```python
plt.figure()
plt.bar(distance, rssi, width=20)
plt.xlabel("Distance [cm]")
plt.ylabel("RSSI [dBm]")
plt.title("RSSI vs Distance")
plt.grid(axis='y')
plt.xticks(distance)
plt.show()
```

Data Rate-Distance

```python
plt.figure()
plt.bar(distance, rate, width=20)
plt.xlabel("Distance [cm]")
plt.ylabel("Data Rate [bit/s]")
plt.title("Data Rate vs Distance")
plt.grid(axis='y')
plt.xticks(distance)
plt.show()
```

### RSSI-BER with Distance markers

```python
plt.figure()
plt.scatter(rssi, ber)

for i in range(len(distance)):
    plt.text(rssi[i], ber[i], f"{distance[i]}")

plt.xlabel("RSSI [dBm]")
plt.ylabel("BER [%]")
plt.title("BER vs RSSI (distance labeled)")
plt.grid()

plt.show()
```
